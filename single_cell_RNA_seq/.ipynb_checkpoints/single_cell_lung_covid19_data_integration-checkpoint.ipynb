## A molecular single-cell lung atlas of lethal COVID-19
- referene: https://www.nature.com/articles/s41586-021-03569-1
- platform: Illumina NovaSeq 6000 and snRNA-seq using a droplet-based platform (10x Genomics)
- overall design: Single-nuclei RNA sequencing of 116,314 cells from 20 frozen lungs obtained from 19 COVID-19 decedents and seven control patients.
- control: 7 patients vs disease: 19 covid-19 patients

```python
import scanpy as sc

import os
import pandas as pd
import numpy as np
```

```python
def sample_preprocessing(sample , input_file):

    ## load the data
    adata = sc.read_csv(input_file).T

    ## filter data on cells and genes
    sc.pp.filter_cells(adata, min_genes=200) ## filter out barcodes(cells) with less than 200 genes
    sc.pp.filter_genes(adata, min_cells=3)   ## filter out genes expressed in less than 3 barcodes(cells)

    ## remove mitochondrial and/or ribosomal genes
    adata.var['mt'] = adata.var_names.str.startswith('MT-') ## label mitochondrial genes
    adata.var['ribosomal'] = adata.var_names.str.startswith(("RPL", "RPS")) ## label ribosomal genes

    ## calculate matrics 
    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)

    ## remove barcode with high gene counts and low gene count using quantile
    upper_quantile = np.quantile(adata.obs.n_genes_by_counts.values, 0.98)
    lower_quantile = np.quantile(adata.obs.n_genes_by_counts.values, 0.02)
    adata = adata[ (adata.obs.n_genes_by_counts < upper_quantile) & (adata.obs.n_genes_by_counts > lower_quantile) ]

    ## remove cells/barcodes with more than 20% mitochondrial genes
    adata = adata[adata.obs.pct_counts_mt < 20 ]

    ## Normalize the data
    sc.pp.normalize_total(adata, target_sum=1e4) ## normalize data to 10,000 UMI counts
    sc.pp.log1p(adata) ## chgange to log counts

    ## get highly variable genes
    sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5) ## defalut parameters

    # save the normalize data before further filtering
    adata.raw = adata

    ## filter data for highly variables genes
    adata = adata[:, adata.var.highly_variable]

    ## regress out effects of total counts and % mitochondial genes
    sc.pp.regress_out(adata, ["total_counts", "pct_counts_mt"])

    ## PCA
    sc.pp.scale(adata, max_value=10) # scale data to unit variance and zero mean
    sc.tl.pca(adata, svd_solver='arpack')

    ## UMAP/leiden
    sc.pp.neighbors(adata, n_neighbors=10, n_pcs=20)
    sc.tl.leiden(adata, resolution=0.25)
    sc.tl.umap(adata)

    return adata
    
```

```python
data_dir = '/home/bonny/Documents/bioinformatics_workspace/datasets/single_cell_pipline/covid-19/'
data_dir
```

```python
sample_dict = dict()

for fi in sorted( os.listdir(data_dir) ):
    file_ext = os.path.splitext(fi)[-1]
    if file_ext == '.gz':
        sample = fi.split('_')[1]
        sample_dict[sample] = data_dir + fi

# sample_dict
```

```python
selected_samples = ['C51ctr', 'C52ctr']
```

```python
out = list()

for sample in selected_samples:
    if sample_dict.get(sample):
        file_path = sample_dict[sample]
        out.append(sample_preprocessing(sample , file_path))
```

```python
lung1, lung2 = out
```
