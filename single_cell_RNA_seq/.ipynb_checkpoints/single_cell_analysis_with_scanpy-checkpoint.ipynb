```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
# load the data 10x genomics PBMC 3k
data_dir = '/home/bonny/Documents/bioinformatics_workspace/datasets/single_cell_pipline/pbmc3k_10x_genomics/filtered_gene_bc_matrices/hg19/'
```

```python
adata = sc.read_10x_mtx(path = data_dir,            # directory with the '.mtx' file located
                       var_names = 'gene_symbols',   # defaut - use gene symbols for variable names
                       cache = True
                       )
```

```python
adata
```

# Preprocessing

### 1) Filter on genes and cells

```python
sc.pp.filter_cells(adata, min_genes=200) ## remove the cells with less than 200 genes.
sc.pp.filter_genes(adata, min_cells=3)   ## remove genes express in fewer the 3 cells.
```

```python
adata
```

### 2) Identify mitochondial genes

```python
adata.var['mt'] = adata.var_names.str.startswith('MT-') ## create column 'mt' identifying mitochondrial genes.
```

```python
adata.var
```

### 3) calculate and plot QC metrics

```python
sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
adata.obs
```

```python
## plot QC metrics - violin plot
sc.pl.violin(adata, ['total_counts', 'n_genes_by_counts', 'pct_counts_mt'], jitter=0.4, multi_panel=True)
```

```python
## plot QC metrics - scatter plot
sc.pl.scatter(adata, x='total_counts', y='n_genes_by_counts')
sc.pl.scatter(adata, x='total_counts', y='pct_counts_mt')
```

```python
## calculate quntile
upper_limit = np.quantile(adata.obs.n_genes_by_counts, 0.98)
lower_limit = np.quantile(adata.obs.n_genes_by_counts, 0.02)

print(f'lower: {lower_limit}, upper:{upper_limit}')
```

### 4) filter data on qc metrics

```python
adata = adata[ (adata.obs.n_genes_by_counts >  lower_limit) & (adata.obs.n_genes_by_counts < upper_limit) ]
```

```python
adata
```

```python
adata = adata[adata.obs.pct_counts_mt < 20, : ]
```

```python
adata
```

# Normalization

```python
sc.pp.normalize_total(adata, target_sum=1e4) ## normalize all the cells to 10,000 UMI
sc.pp.log1p(adata)                           ## apply log to the normalized counts data
```

```python
adata
```

# Clustering 

```python
## get highly variable genes (default parameter applied)
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
adata.var[adata.var.highly_variable]
```

```python
## filter on highly variable genes
adata = adata[:, adata.var.highly_variable]
adata
```

```python
## regress out effect of total counts per cell and percent mt 
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt'])
```

```python
## scale data to unit variance and zero mean
sc.pp.scale(adata, max_value=10)
```

```python
## apply PCA
sc.tl.pca(adata, svd_solver='arpack')
```

```python
## plot PCA
sc.pl.pca_variance_ratio(adata, log=True)
```

```python
## caclulate neighbor for UMAP
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=15)
```

```python
sc.tl.umap(adata)
sc.pl.umap(adata)
```

```python
sc.tl.leiden(adata)
sc.pl.umap(adata, color=['leiden'])
```

# Find Markers

```python
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=15, sharey=False)
```

```python
adata
```

```python
adata.uns['rank_genes_groups']['names']
```

```python
out = np.array([[0,0,0,0,0]])
for group in adata.uns['rank_genes_groups']['names'].dtype.names:
    out = np.vstack((out,
                   np.vstack((adata.uns['rank_genes_groups']['names'][group],
                             adata.uns['rank_genes_groups']['scores'][group],
                             adata.uns['rank_genes_groups']['pvals_adj'][group],
                             adata.uns['rank_genes_groups']['logfoldchanges'][group],
                             np.array([group] * len(adata.uns['rank_genes_groups']['names'][group])).astype('object'))).T))
                   
```

```python
out.shape
```

```python
markers = pd.DataFrame(out[1:], columns=['Genes','scores','pval_adj','lfc','cluster'])
```
